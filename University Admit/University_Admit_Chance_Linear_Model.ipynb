```python
X = admin.iloc[:,1:8]
y = admin.iloc[:,8]
```

```python
#Sampling
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = 101)
```

```python
#Fit model to dataset
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(X_train,y_train)
```

```python
pred_chance = linear.predict(X_test)
pred_chance
```

```python
chance_df = dict({"Actual Chance":y_test,"Predicted Chance":pred_chance})
chance_df = pd.DataFrame(chance_df)
chance_df
```

```python
linear.score(X_train,y_train)  #0.7981798784416266
```

```python
from sklearn import metrics
print("The Mean Squared Error is : ",metrics.mean_squared_error(y_test,pred_chance) )
print("The Mean Absolute Error is : ",metrics.mean_absolute_error(y_test,pred_chance) )
print("The Root Mean Squared Error is : ",pow(metrics.mean_absolute_error(y_test,pred_chance) ,0.5) )
```
